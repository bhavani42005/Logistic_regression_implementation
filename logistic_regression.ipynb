```python
import numpy as np
```

```python
class LogisticRegression:
    def __init__(self,learning_rate=0.01,epochs=5000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bais = None
    def sigmoid(self,z):
        return 1/(1+np.exp(-z))
    def fit(self,x,y):
        self.n_features = x.shape[1]
        n_instances = x.shape[0]
        self.weights = np.zeros(self.n_features)
        self.bais = 0
        lr = self.learning_rate
        for _ in range(self.epochs):
            
            y_output = np.dot(x,self.weights) + self.bais
            d_loss = (1/n_instances)*(np.dot(x.T,self.sigmoid(y_output)-y))
            d_bais = (1/n_instances)*(np.sum(self.sigmoid(y_output)-y))
            predictions = np.clip(self.sigmoid(y_output),1e-15,1-1e-15)
            loss = -np.mean(y*np.log(predictions)+(1-y)*np.log(1-predictions))
            if _ %100 == 0 :
                print(loss)
            self.weights -= d_loss*lr
            self.bais -= d_bais*lr
            if _ == 0 :            
                print("predictions",predictions)
                print("weights",self.weights)
                print("bais",self.bais)
                new_output = np.dot(x, self.weights) + self.bais
                print("Min z:", np.min(new_output))
                print("Max z:", np.max(new_output))
    def predict_proba(self,x):
        y_out = np.dot(x,self.weights)+self.bais
        return self.sigmoid(y_out)
    def predict(self,x):
       return np.where(self.predict_proba(x) >= 0.5,1,0)
```

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\bhava\Downloads\studyvspassdata\marks.csv")
data.head()
```

```python
data = data.reset_index()
data.head()
```

```python
from zlib import crc32
def test_set_check(identifier,test_ratio):
    return crc32(np.int64(identifier))& 0xFFFFFFFF< test_ratio*2**32
    
```

```python
def split_data_id(test_ratio,id_column):
    id = data[id_column]
    in_test_set = id.apply(lambda id_:test_set_check(id_,test_ratio))
    return data.loc[~in_test_set],data.loc[in_test_set]
    
```

```python
train,test = split_data_id(0.1,"index")
```

```python
train_y = train["result"]
train_x = train.drop("result",axis=1)
train_x = train_x.drop("index",axis=1)
test_y = test["result"]
test_x = test.drop("result",axis=1)
test_x = test_x.drop("index",axis=1)
```

```python
train_x.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
pipe = Pipeline([
    ("imputer",SimpleImputer(strategy = "median")) ,
("std_scaler",StandardScaler())])
```

```python
train_x = pd.DataFrame(
    pipe.fit_transform(train_x),
    columns=train_x.columns,
    
)

test_x = pd.DataFrame(
    pipe.transform(test_x),
    columns=test_x.columns,
   
)
```

```python
print(train_x)
```

```python

model = LogisticRegression()
model.fit(train_x,train_y)

```

```python
test_x = test_x.drop("index",axis=1,errors="ignore")
y_test_pred = model.predict(test_x)
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score,recall_score
precision_score(y_test_pred,test_y)
```

```python
f1_score(y_test_pred,test_y)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test_pred,test_y)
```

```python
from sklearn.linear_model import LogisticRegression
m1 = LogisticRegression(max_iter=1000)
m1.fit(train_x,train_y)
res = m1.predict(test_x)
accuracy_score(res,test_y)
```
